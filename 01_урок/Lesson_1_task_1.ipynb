Постройте на одном графике две кривые y(x) для функции двух переменной y(k,x)=cos(k∙x), взяв для одной кривой значение k=1, а для другой – любое другое k, не равное 1.

```python
import numpy as np
import plotly.plotly as py
import plotly.graph_objs as go
import matplotlib.pyplot as plt
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
```

```python
init_notebook_mode(connected=True)
```

```python
x = np.linspace(-np.pi, np.pi, 201)
```

```python
k_1 = 1
k_2 = int(input('Введите значение k  '))
```

```python
plt.plot(x, np.cos(k_1*x))
plt.plot(x, np.cos(k_2*x));
```

```python
a = go.Scatter(name='cos({}*x)'.format(k_1),
                   x = x, y = np.cos(k_1*x))
b = go.Scatter(name='cos({}*x)'.format(k_2),
                   x = x, y = np.cos(k_2*x))
```

```python
iplot([a, b])
```
